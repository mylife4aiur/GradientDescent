```python
plota = np.arange(-4, 4, 0.25)
plotb = np.arange(-4, 4, 0.25)
plota, plotb = np.meshgrid(plota, plotb)
```

```python
def PlotJ(train,a,b):
    J=np.zeros(np.shape(a))
    for i in train:
        Z_pre=np.array((a*i[0]**2 + b*i[1]**2))
        Z = np.array(2*i[0]**2 -1*i[1]**2)
        J+=(Z_pre-Z)**2
    return J
```

```python
plotJ=PlotJ(train,plota,plotb)
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)

# 具体函数方法可用 help(function) 查看，如：help(ax.plot_surface)
ax.plot_surface(plota, plotb, plotJ, rstride=1, cstride=1, cmap='rainbow',alpha=0.75)
plt.show()
```

```python
from itertools import combinations
train=list(combinations([0.1*i for i in range(10)], 2))
```

```python
def dJ(train,a,b):
    tmp=a*train[0]**2+b*train[1]**2-(2*train[0]**2-train[1]**2)
    djda=tmp*train[0]**2
    djdb=tmp*train[1]**2
    return djda,djdb
```

```python
def J(train,a,b):

    Z_pre=np.array((a*train[0]**2 + b*train[1]**2))
    Z = np.array(2*train[0]**2 -1*train[1]**2)
    J =(Z_pre-Z)**2
    return J
```

```python
pre_a=100
pre_b=100
a=4
b=4
alpha=1
iteration=0

x=[]
y=[]
z=[]

while(((abs(a-pre_a)>1e-6) and abs(b-pre_b)>1e-6) and iteration<1000):
    iteration+=1
    J_val=0
    djda=0
    djdb=0
    for i in train:
        J_val+=J(i,a,b)
        tmp1,tmp2=dJ(i,a,b)
        djda+=tmp1
        djdb+=tmp2
    djda=djda/len(train)
    djdb=djdb/len(train)
    
    x.append(a)
    y.append(b)
    z.append(J_val)
    
    pre_a=a
    pre_b=b
    a-=alpha*djda
    b-=alpha*djdb
    
    
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm # sklearn = scikit-learn
from sklearn.datasets import make_moons
from moviepy.editor import VideoClip
from moviepy.video.io.bindings import mplfig_to_npimage

X, Y = make_moons(50, noise=0.1, random_state=2) # semi-random data

fig, ax = plt.subplots(1, figsize=(4, 4), facecolor=(1,1,1))
fig.subplots_adjust(left=0, right=1, bottom=0)
xx, yy = np.meshgrid(np.linspace(-2,3,500), np.linspace(-1,2,500))

def make_frame(t):
    ax.clear()
    ax.axis('off')
    ax.set_title("SVC classification", fontsize=16)

    classifier = svm.SVC(gamma=2, C=1)
    # the varying weights make the points appear one after the other
    weights = np.minimum(1, np.maximum(0, t**2+10-np.arange(50)))
    classifier.fit(X, Y, sample_weight=weights)
    Z = classifier.decision_function(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    ax.contourf(xx, yy, Z, cmap=plt.cm.bone, alpha=0.8,
                vmin=-2.5, vmax=2.5, levels=np.linspace(-2,2,20))
    ax.scatter(X[:,0], X[:,1], c=Y, s=50*weights, cmap=plt.cm.bone)

    return mplfig_to_npimage(fig)

animation = VideoClip(make_frame, duration = 7)
animation.write_gif("svm.gif", fps=15)
```
